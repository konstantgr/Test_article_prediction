```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score

path='data/predict.csv'
```

## Read data

```python
df_raw = pd.read_csv(path)
df_raw.fillna(0, inplace=True)
df_raw.head(10)
```

## Preprocessing

```python
df_raw.describe()
```

```python
df = df_raw.drop(['authors_mean_pagerank', 'authors_mean_productivity',
                 'journal_pagerank'], 1)
```

Посмотрим, как от различных параметров зависит <i>"c5"</i>

```python
feautures = list(df)
feautures.remove('c5')

plots_in_row = 4
for row in range(3):
    start = row * plots_in_row
    stop = (row + 1) * plots_in_row
    sns.pairplot(df, y_vars='c5', x_vars=feautures[start:stop],
                 palette='spring')
```

## Predictions

```python
def LR(X, y, test_size) -> tuple:
    ''' Linear regression model.

    Returns
    -------
    tuple
        a tuple of following elements:

        y_test : np.ndarray 
            Test sample 
        predicted : np.ndarray
            Predicted values using linear regression model
        r2 : float    
            Value of coefficient of determination
    '''
    
    (X_train, X_test, y_train, y_test) = train_test_split(X, y,
                                                          test_size=test_size, 
                                                          random_state=10)
    
    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    predicted = np.floor(y_pred)
    r2 = r2_score(y_test, predicted)
    return (y_test, predicted, r2)


def GBR(X, y, test_size=0.4, n_estimators=100, learning_rate=0.1):
    ''' Gradient boosting regression model. 
    
    Returns
    -------
    tuple
        a tuple of following elements:
        
        y_test : np.ndarray 
            Test sample 
        predicted : np.ndarray
            Predicted values using gradient boosting regression model
        r2 : float    
            Value of coefficient of determination
    '''
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                        test_size=test_size, 
                                                        random_state=10)

    model = GradientBoostingRegressor(random_state=32, 
                                      n_estimators=n_estimators, 
                                      learning_rate=learning_rate)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    predicted = np.floor(y_pred)
    return (y_test, predicted, r2_score(y_test, predicted))


def RD(X, y, test_size):
    ''' Ridge regression model. 
    
    Returns
    -------
    tuple
        a tuple of following elements:
        
        y_test : np.ndarray 
            Test sample 
        predicted : np.ndarray
            Predicted values using ridge regression model
        r2 : float    
            Value of coefficient of determination
    '''
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                        test_size=test_size, 
                                                        random_state=10)
    
    model = Ridge(alpha=1e-2)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    predicted = np.floor(y_pred)
    return (y_test, predicted, r2_score(y_test, predicted))
```

Посмотрим, какой результат выдаст обучение модели алгоритмом градиентного бустинга.

```python
X = df.drop(['c5', 'authors_mean_hindex', 'authors_mean_gindex'],
            axis=1).values
y = df.c5.values

y_test, y_pred, r2 = GBR(X, y, test_size=0.4)

plt.figure(figsize=(5, 5))
plt.text(0, 50, "R^2 = {:.2f}".format(r2), 
         fontsize=14)
plt.scatter(y_pred, y_test, s=10, color='red')
plt.xlabel('Predicted citations', fontsize=14)
plt.ylabel('Real citations', fontsize=14)
plt.show()
```

Можно заметить, что даже неоптимизированный градиентный бустинг уже выдает довольно неплохой результат. Сравним его с другими регрессионными меоделями на примере линейной регресии и гребневой регресии. Будем искать R^2 в зависимости от соотношения тестовой выборки к обучающей.

```python
X = df.drop(['c5', 'authors_mean_hindex', 'authors_mean_gindex'], 
            axis=1).values
y = df.c5.values

test_sizes = np.linspace(0.1, 0.85, 200, endpoint=True)
LR_res = []
GBR_res = []
RD_res = []

for test_size in test_sizes:
    GBR_res.append(GBR(X, y, test_size)[2])
    LR_res.append(LR(X, y, test_size)[2])
    RD_res.append(RD(X, y, test_size)[2])
```

```python
plt.scatter(test_sizes, GBR_res, s=8, label='Gradient boosting regression') 
plt.scatter(test_sizes, LR_res, s=8, c='orange', label='Linear regression')
plt.scatter(test_sizes, RD_res, s=8, c='green', label='Ridge regression')  
plt.hlines(0.7, 0, 1, linestyle='--', color='red', linewidth=1)
plt.vlines(0.4, 0, 1, linestyle='--', color='red', linewidth=1)

plt.ylim(0, 1)
plt.xlim(0, 1)
plt.xlabel('Test size ratio', fontsize=14)
plt.ylabel('R^2', fontsize=14)
plt.yticks(list(plt.yticks()[0]) + [0.7])
plt.legend()
plt.show()
```

Из графика видно, что лучше всего себя показал, как неудивительно, градиентный бустинг.<br> 
Постараемся улучшить результат градинетного бустинга, оптимизировав его параметры.

```python
number_of_points = 20
n_estimators = np.linspace(10, 60, number_of_points, dtype=int, endpoint=True)
learning_rates = np.linspace(0.05, 0.85, number_of_points, endpoint=True)
r2_arr = []; estimators = []; rates = []

for n in n_estimators:
    for learning_rate in learning_rates:
            estimators.append(n)
            rates.append(learning_rate)
            r2_arr.append(GBR(X, y, test_size=0.4, 
                              n_estimators=n, 
                              learning_rate=learning_rate)[2])
r2_arr = np.array(r2_arr)
rates = np.array(rates)
estimators = np.array(estimators)

boosting_df = pd.DataFrame(
    data=r2_arr.reshape(number_of_points, number_of_points),
    columns=rates.reshape(number_of_points, number_of_points)[0].round(2),
    index=estimators.reshape(number_of_points, number_of_points).T[0])

sns.heatmap(boosting_df, square=True)
plt.show()
```

Видно, что от n_estimators происходит быстрое насыщение и самое эффективное значение learning rate находится в районе 0.43, при котором R^2 равен уже <b>0.8</b>!

```python
X = df.drop(['c5', 'authors_mean_hindex', 'authors_mean_gindex'], 
            axis=1).values
y = df.c5.values

y_test, y_pred, r2 = GBR(X, y, test_size=0.4, 
                         n_estimators=33, 
                         learning_rate=0.43)

plt.figure(figsize=(5, 5))
plt.text(0, 45, "R^2 = {:.2f}".format(r2), 
         fontsize=14)
plt.scatter(y_pred, y_test, s=10, color='red')
plt.xlabel('Predicted citations', fontsize=14)
plt.ylabel('Real citations', fontsize=14)
plt.show()
```

```python
df_predictions = pd.DataFrame({'Real': y_test, 'Predicted': y_pred})
df_predictions.head(10)
```

Таким образом, с помощью градиентного бустинга получили R^2 равный <b>0.81</b>. В итоге, можно сказать, что даже такими базовыми алгоритмами получилось достичь высокой точности предсказаний.
